```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

VIDEOS = pd.read_csv("/kaggle/input/datathon-loreal/videos.csv")
HAWKES = pd.read_csv("/kaggle/input/apply-hawkes-process-for-self-exciting-momentum/signal_hawkes.csv")
TBI = pd.read_csv("/kaggle/input/tbi-novelty-detection/tbi_novelty_bursts.csv")
FUNDAMENTAL = pd.read_csv("/kaggle/input/fundamental-analysis-comments-likes-tags/signal_fundamentals.csv")
DECAY = pd.read_csv("/kaggle/input/trend-decay-analysis-multi-signal-combined/signal_decay.csv")
FUSION = pd.read_csv("/kaggle/input/fusion-engine/all_signals_combined.csv")

print("/kaggle/input/datathon-loreal/videos.csv: ")
print("Shape:", VIDEOS.shape)
display(VIDEOS.head())

print("/kaggle/input/apply-hawkes-process-for-self-exciting-momentum/signal_hawkes.csv: ")
print("Shape:", HAWKES.shape)
display(HAWKES.head())

print("/kaggle/input/tbi-novelty-detection/tbi_novelty_bursts.csv: ")
print("Shape:", TBI.shape)
display(TBI.head())

print("/kaggle/input/fundamental-analysis-comments-likes-tags/signal_fundamentals.csv: ")
print("Shape:", FUNDAMENTAL.shape)
display(FUNDAMENTAL.head())

print("/kaggle/input/trend-decay-analysis-multi-signal-combined/signal_decay.csv: ")
print("Shape:", DECAY.shape)
display(DECAY.head())

print("/kaggle/input/fusion-engine/all_signals_combined.csv: ")
print("Shape:", FUSION.shape)
display(FUSION.head())
```

```python
# Let's examine the column names and structure of each dataset
print("VIDEOS columns:")
print(VIDEOS.columns.tolist())
print("\nHAWKES columns:")
print(HAWKES.columns.tolist())
print("\nTBI columns:")
print(TBI.columns.tolist())
print("\nFUNDAMENTAL columns:")
print(FUNDAMENTAL.columns.tolist())
print("\nDECAY columns:")
print(DECAY.columns.tolist())
print("\nFUSION columns:")
print(FUSION.columns.tolist())

# Check for common key columns (likely videoId)
print("\n" + "="*50)
print("Checking for videoId in each dataset:")
for name, df in [("VIDEOS", VIDEOS), ("HAWKES", HAWKES), ("TBI", TBI), 
                 ("FUNDAMENTAL", FUNDAMENTAL), ("DECAY", DECAY), ("FUSION", FUSION)]:
    has_videoid = 'videoId' in df.columns
    print(f"{name}: {'Yes' if has_videoid else 'No'}")
    if has_videoid:
        print(f"  - videoId range: {df['videoId'].min()} to {df['videoId'].max()}")
        print(f"  - Unique videoIds: {df['videoId'].nunique()}")
    print()
```

```python
# Create the master dataset by merging all signals with video metadata
# Start with the VIDEOS dataset as it contains titles, descriptions, and tags

# First, let's prepare the VIDEOS dataset by selecting key columns
video_metadata = VIDEOS[['videoId', 'title', 'description', 'tags', 'publishedAt', 
                        'channelId', 'contentDuration', 'viewCount', 'likeCount', 
                        'commentCount', 'topicCategories']].copy()

print(f"Starting with video metadata: {video_metadata.shape}")
print(f"VideoId range: {video_metadata['videoId'].min()} to {video_metadata['videoId'].max()}")

# Convert videoId to consistent type for merging
video_metadata['videoId'] = video_metadata['videoId'].astype(int)

# Now let's merge with each signal dataset
# Start with the largest signal dataset (FUSION) as the base
master_df = FUSION.copy()
master_df['videoId'] = master_df['videoId'].astype(int)

print(f"\nStarting with FUSION dataset: {master_df.shape}")

# Merge with video metadata
master_df = master_df.merge(video_metadata, on='videoId', how='left')
print(f"After merging with video metadata: {master_df.shape}")

# Merge with FUNDAMENTAL data (add prefix to avoid column conflicts)
fundamental_cols = FUNDAMENTAL.columns.tolist()
fundamental_cols.remove('videoId')  # Remove videoId to avoid duplication
fundamental_renamed = FUNDAMENTAL.rename(columns={col: f'fund_{col}' for col in fundamental_cols})
fundamental_renamed['videoId'] = fundamental_renamed['videoId'].astype(int)

master_df = master_df.merge(fundamental_renamed, on='videoId', how='left')
print(f"After merging with FUNDAMENTAL: {master_df.shape}")

# Merge with DECAY data
decay_renamed = DECAY.copy()
decay_renamed['videoId'] = decay_renamed['videoId'].astype(int)
decay_cols = [col for col in decay_renamed.columns if col != 'videoId']
decay_renamed = decay_renamed.rename(columns={col: f'decay_{col}' for col in decay_cols})

master_df = master_df.merge(decay_renamed, on='videoId', how='left')
print(f"After merging with DECAY: {master_df.shape}")

# Merge with HAWKES data (add prefix to avoid conflicts)
hawkes_renamed = HAWKES.copy()
hawkes_renamed['videoId'] = hawkes_renamed['videoId'].astype(int)
hawkes_cols = [col for col in hawkes_renamed.columns if col != 'videoId']
hawkes_renamed = hawkes_renamed.rename(columns={col: f'hawkes_{col}' for col in hawkes_cols})

master_df = master_df.merge(hawkes_renamed, on='videoId', how='left')
print(f"After merging with HAWKES: {master_df.shape}")

# Merge with TBI data (add prefix to avoid conflicts)
tbi_renamed = TBI.copy()
tbi_renamed['videoId'] = tbi_renamed['videoId'].astype(int)
tbi_cols = [col for col in tbi_renamed.columns if col != 'videoId']
tbi_renamed = tbi_renamed.rename(columns={col: f'tbi_{col}' for col in tbi_cols})

master_df = master_df.merge(tbi_renamed, on='videoId', how='left')
print(f"After merging with TBI: {master_df.shape}")

print(f"\nFinal master dataset shape: {master_df.shape}")
print(f"Total columns: {len(master_df.columns)}")
print("\nColumn summary:")
print("- Video metadata: title, description, tags, publishedAt, channelId, etc.")
print("- FUSION signals: composite_score, performance_category, predicted_future_score, etc.")
print("- FUNDAMENTAL signals: fund_total_comments, fund_engagement_ratio, fund_fundamental_health, etc.")
print("- DECAY signals: decay_strength_combined, decay_half_life_hours_combined, decay_state_combined")
print("- HAWKES signals: hawkes_momentum_ratio, hawkes_composite_momentum, hawkes_velocity, etc.")
print("- TBI signals: tbi_novelty_strength, tbi_novelty_duration, tbi_avg_tbi, etc.")
```

```python
# Preview the master dataset
print("Master Dataset Preview:")
print("=" * 50)
print(f"Shape: {master_df.shape}")
print(f"Memory usage: {master_df.memory_usage(deep=True).sum() / 1024**2:.2f} MB")

# Show first few rows with key columns
key_columns = ['videoId', 'title', 'description', 'tags', 'composite_score', 
               'performance_category', 'predicted_future_score']
print(f"\nPreview of key columns:")
display(master_df[key_columns].head())

# Show column names organized by category
print(f"\nAll {len(master_df.columns)} columns:")
print("\n1. Core Video Info:")
core_cols = [col for col in master_df.columns if col in ['videoId', 'title', 'description', 'tags', 
             'publishedAt', 'channelId', 'contentDuration', 'viewCount', 'likeCount', 'commentCount', 'topicCategories']]
print(core_cols)

print("\n2. FUSION Signals:")
fusion_cols = [col for col in master_df.columns if not any(col.startswith(prefix) for prefix in ['fund_', 'decay_', 'hawkes_', 'tbi_']) 
               and col not in core_cols]
print(fusion_cols)

print("\n3. FUNDAMENTAL Signals:")
fund_cols = [col for col in master_df.columns if col.startswith('fund_')]
print(fund_cols)

print("\n4. DECAY Signals:")
decay_cols = [col for col in master_df.columns if col.startswith('decay_')]
print(decay_cols)

print("\n5. HAWKES Signals:")
hawkes_cols = [col for col in master_df.columns if col.startswith('hawkes_')]
print(hawkes_cols)

print("\n6. TBI Signals:")
tbi_cols = [col for col in master_df.columns if col.startswith('tbi_')]
print(tbi_cols)
```

```python
# Save the master dataset to CSV
output_file = "master.csv"

print("Saving master dataset to CSV...")
print(f"File: {output_file}")
print(f"Shape: {master_df.shape}")
print(f"Columns: {len(master_df.columns)}")

# Save to CSV
master_df.to_csv(output_file, index=False)

print(f"\n‚úÖ Successfully saved master dataset to '{output_file}'!")
print(f"üìä Dataset contains {master_df.shape[0]:,} videos with {master_df.shape[1]} features each")
print(f"üìÅ File size: approximately {master_df.memory_usage(deep=True).sum() / 1024**2:.1f} MB")

# Verify the file was created
import os
if os.path.exists(output_file):
    file_size_mb = os.path.getsize(output_file) / 1024**2
    print(f"‚úÖ File verification: {output_file} created successfully ({file_size_mb:.1f} MB)")
else:
    print("‚ùå Error: File was not created successfully")

# Show a summary of what's included
print(f"\nüìã Master dataset includes:")
print(f"   ‚Ä¢ Video metadata: titles, descriptions, tags, publication info")
print(f"   ‚Ä¢ FUSION signals: composite scores, performance categories, predictions")
print(f"   ‚Ä¢ FUNDAMENTAL signals: engagement metrics, content analysis")
print(f"   ‚Ä¢ DECAY signals: trend decay analysis")
print(f"   ‚Ä¢ HAWKES signals: momentum and self-exciting processes")
print(f"   ‚Ä¢ TBI signals: novelty detection and burst analysis")
```
